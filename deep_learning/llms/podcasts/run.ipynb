# Evaluate

```julia
from transformers import TFAutoModelForCausalLM, AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
tokenizer.pad_token = tokenizer.eos_token
base_model = TFAutoModelForCausalLM.from_pretrained("distilgpt2", pad_token_id=tokenizer.eos_token_id)

def get_response(query, model, tokenizer):
    input_ids = tokenizer.encode(query,add_special_tokens=False, return_tensors="tf")
    decoder_start_token_id = input_ids[0][-1]

    output = model.generate(
        input_ids,
        min_length=50,
        max_length=len(input_ids[0]) + 4096,
        top_k=40,
        top_p=0.9,
        no_repeat_ngram_size=2,
        decoder_start_token_id=decoder_start_token_id,
        early_stopping=False
    )
    return tokenizer.decode(output[0][len(input_ids[0]):], skip_special_tokens=True)
```

```julia
from warnings import filterwarnings
filterwarnings("ignore")

class Color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```julia
#@title Question
input_query = "What is the meaning of life?" #@param {type:"string"}
print(f"{Color.BOLD}{Color.RED}Question: {input_query}{Color.END}\n")
print(f"{Color.BOLD}{Color.DARKCYAN}Baseline Bot:{get_response(input_query, base_model, tokenizer)}{Color.END}\n")
print(f"{Color.BOLD}{Color.PURPLE}Casper Bot:{get_response(input_query, model, tokenizer)}{Color.END}")
```
